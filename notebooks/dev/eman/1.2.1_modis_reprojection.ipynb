The very first thing we notice is that our coordinates (lat-lon) are 2D.

This is very troublesome for a lot of software because it's not regular.

## Reprojection: Regular Grid

The first option is to reproject the data onto a regular grid.
In this case, regular means Lat-Lon but with a constant `dlon,dlat`.
This behaves like an image which makes it easier to work with for tools.

```python
%%time
# regrid to regular
resolution = 0.01
calibration = "radiance"
modis_grid = regrid_swath_to_regular(modis_swath, calibration=calibration, resolution=resolution)
```

Now, we can look at the coordinates.
We should have nice 1D lat-lon coordinates.
This works more like an image where each dlon, dlat is equal in length (units are degrees).

```python
# 
modis_grid.coords
```

In addition, calculating the footprint (polygon) will be easier!

```python
modis_grid_polygon = calculate_xrio_footprint(modis_grid)

modis_grid_polygon
```

...and right away, we notice that it is much more "square".

```python
# in an even better way 
fig = plt.figure()
ax = plt.axes(projection=ccrs.PlateCarree())

modis_grid.isel(band=0)[calibration].plot(
    ax=ax,
    x="longitude", y="latitude", 
    transform=ccrs.PlateCarree(),
)

ax.add_geometries(
    [modis_grid_polygon],
    crs=ccrs.CRS(modis_grid.rio.crs),
    facecolor="none",
    edgecolor="tab:green",
    alpha=1,
    linewidth=5,
    zorder=10,
)

ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())

ax.coastlines()
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=0.1, color='k', alpha=1, 
                  linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size': 8}
gl.ylabel_style = {'size': 8}
plt.tight_layout()
plt.show()
```

#### Warning - Warping

Be very careful because there could be some induced warping that occurs.
This is natural because we are going from a 2D curvilinear grid to a 1D regular grid.

```python
# in an even better way 
fig = plt.figure()
ax = plt.axes(projection=ccrs.PlateCarree())

modis_grid.isel(band=0)[calibration].plot(
    ax=ax,
    x="longitude", y="latitude", 
    transform=ccrs.PlateCarree(),
)

ax.add_geometries(
    [modis_grid_polygon],
    crs=ccrs.CRS(modis_grid.rio.crs),
    facecolor="none",
    edgecolor="tab:green",
    alpha=1,
    linewidth=10,
    zorder=10,
)

ax.coastlines()
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=0.1, color='k', alpha=1, 
                  linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size': 8}
gl.ylabel_style = {'size': 8}
plt.tight_layout()
plt.show()
```

### Key: CRS Information

There are some key pieces of information that are absolutely necessary whenever we need to do some transformations:

* coordinate reference system (CRS)
* bounds
* resolution
* transformation

With these 4 pieces of information, one can virtually move between **any** coordinate reference system (and back).
Traditionally, in python, one would need a few packages including `rasterio`, `shapely`, and `affine`.
Fortunately, the `rioxarray` package allow us to use this functionaly with native `xarray`.
You just need to access the `accessor` (i.e., `xr.Dataset.rio...`) and you can access all of the relevant information.

```python
# CRS
src_crs = modis_grid.rio.crs
print(f"CRS:\n{src_crs}\n")

# bounds
src_bounds = modis_grid.rio.bounds()
print(f"Bounds:\n{src_bounds}\n")

# resolution
src_resolution = modis_grid.rio.resolution()
print(f"Resolution:\n{src_resolution}\n")

# transformation
src_transform = modis_grid.rio.transform()
print(f"Transform:\n{src_transform}\n")
```

### Reprojection: UTM

Sometimes, we want a nice image with some measurements that are measured in meters.
This works nicely because it will transform the data into meters instead of degrees which is much easier to manipulate.

```python
# estimate utm
utm_espg = modis_grid.rio.estimate_utm_crs()
print(f"Approximate UTM Projection: {utm_espg}")
```

```python
%%time

# reproject to utm
resolution = 1_000 # meters
modis_grid_utm = modis_grid.rio.reproject(utm_espg, resolution=resolution)
modis_grid_utm.rio.shape
```

```python
# CRS
src_crs = modis_grid_utm.rio.crs
print(f"CRS:\n{src_crs}\n")

# bounds
src_bounds = modis_grid_utm.rio.bounds()
print(f"Bounds:\n{src_bounds}\n")

# resolution
src_resolution = modis_grid_utm.rio.resolution()
print(f"Resolution:\n{src_resolution}\n")

# transformation
src_transform = modis_grid_utm.rio.transform()
print(f"Transform:\n{src_transform}\n")
```

```python
# calculate footprint (polygon)
modis_grid_utm_polygon = calculate_xrio_footprint(modis_grid_utm)
modis_grid_utm_polygon
```

```python
fig, ax = plt.subplots()
modis_grid_utm.isel(band=0)[calibration].plot.imshow(
    ax=ax,
    x="x", y="y", 
    vmax=500
    # transform=modis_grid_utm.rio.transform,
)

polygon_coords = modis_grid_utm_polygon.exterior.xy
ax.plot(*polygon_coords, linewidth=10, color="tab:green")


plt.tight_layout()
plt.show()
```
