## downloading data using goes2go

```python
from goes2go import GOES

# ABI Level 1b Data
G = GOES(satellite=16, product="ABI-L1b-Rad", domain='F')
```

```python
# download the data nearest a specific time
files = G.nearesttime('2018-10-02 17:00', return_as='filelist')#, save_dir=None)
```

```python
files.file[0]
```
